### [Numpy](https://numpy.org/doc) - package for vectorized operations

Python has nearly unlimited capabilities for Data Science. One such example is **a package for working with vectors and matrices (numpy)**

```python
# now we will explore numpy - best tool for working with arrays and matrices
import numpy as np

# we can create a numpy array from a simple python list (python lists ~ C++ vectors)
lst = [1, 2, 3, 4]  # creating python list
np_arr = np.array(lst)  # creating numpy array
np_arr  # displaying created array
```

We have just created a simple numpy array. Below are several more options on how you can create numpy arrays in other ways for different use-cases:

```python
range_arr = np.arange(1,9,1)  # np.arange(start_value, end_value, step) - end_value is NOT included
print("range_arr", range_arr)

uniform_arr = np.linspace(1,9,9)  # np.linspace(start_value, end_value, number_of_items_in_the_array) - end_value IS included
print("uniform_arr", uniform_arr)

ones_arr = np.ones((2,3))  # np.ones(shape_of_array)
print("ones_arr\n", ones_arr)
```

It should be noted that all operations for numpy arrays are parallelized - for example, if you try to create an array of ones using a for-loop instead of a numpy constructor, it will take you way longer. Here is an example:

```python
from time import time
N = 10000000

# using NumPy package
start = time()
a = np.ones((1,N))
b = np.ones((N,1))
end = time()
print(f"Array creation done in {round(end - start, 2)} sec")
start = time()
c = a @ b
end = time()
print(f"a @ b = {c}, completed in {round(end-start, 2)} sec")

# Using pure Python
start = time()
a = [1 for i in range(N)]
b = [1 for i in range(N)]
end = time()
print(f"\nArray creation done in {round(end - start, 2)} sec")
start = time()
c = 0
for item_a, item_b in zip(a, b):
  c += item_a * item_b
end = time()
print(f"a @ b = {c}, completed in {round(end-start, 2)} sec")
```

The package also has a wide list of vector and matrix operators for array permutations of any type: multiplication, summation, inverse matrix calculation and much more:

```python
# creating a demo array - 3-element vectors
a = np.array([1,2,3])

# testing different functions with a single operand array
b = a**2
print("element-wise power of 2 function", b)

b = np.sin(a)
print("\nelement-wise sin function", b)

b = np.exp(a)
print("\nelement-wise exp function", b)

b = np.sqrt(a)
print("\nelement-wise sqrt function", b)
```

### [Pandas](https://pandas.pydata.org/docs) - tabular data manipulations

Visualizations are the foundations of Data Science. Let us take a look at what we can achieve with Python. We will begin with simple curve plots:

```python
# now goes the vusialization part
import matplotlib.pyplot as plt

# let us create some data to plot
x = np.arange(1,10.1,0.5)
y = x ** 2
```

```python
# now let us create a plot of our square function
plt.plot(x, y)  # plotting the data on our canvas
plt.show()  # displaying the canvas so that we can see it
```

The plot is complete. However, it looks rather poor - let us improve it slightly:

```python
plt.plot(x, y, '-o', color='b')
plt.title("$f(x)=x^2$")  # title can be formatted with LATEX
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

A bit better - let us continue:

```python
x = np.linspace(-5*np.pi,5*np.pi,1000)
y1 = np.sin(x)
y2 = np.cos(x) / 2
plt.plot(x, y1, color='b', label='f(x) = sin(x)')  # plotting first curve
plt.plot(x, y2, '--', color='orange', label='f(x) = cos(x) / 2')  # plotting second curve
plt.legend()
plt.xlabel("X")
plt.ylabel("f(x)")
plt.title("Some demo plot")
plt.show()
```

As we can see, we can add several lines of different styles at once, creating compact but highly informative plots. Just remember to keep it comprehensible and not too overloaded.

Our dataframe contains a bunch of points for creating 13 pictures. Let us visualize the first 12 of them:

```python
df['dataset'].unique().tolist()
```

```python
fig, axs = plt.subplots(3, 4, figsize=(12, 8))
subsets = df['dataset'].unique().tolist()[:12]
c = -1

# Iterate over each subplot in the 3x4 grid
for i in range(3):
    for j in range(4):
        c += 1
        data = df[df['dataset']==subsets[c]]
        ax = axs[i, j]
        ax.scatter(data['x'], data['y'], s=5, c="r")
        ax.set_title(f'Subset {subsets[c]}')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
